```python
plt.figure(figsize=(12,4))
plt.plot(data_2['General index'], label = 'General index')
plt.legend(loc = 'best')
plt.title('distribution of CPI')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(data_2['General index'])
print(adf_test)
print('ADF statistic: %f' % adf_test[0])
print('Critical Values @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

```python
from scipy.stats import boxcox
data_boxcox = pd.Series(boxcox(data_2['General index'], lmbda = 0), index = data_2.index)
plt.figure(figsize=(12,4))
plt.plot(data_boxcox, label = 'after boxcox transformation')
plt.legend(loc='best')
plt.title('After boxcox transformation\n', fontdict = {'fontsize':16, 'fontweight':5, 'color' : 'Orange'})
plt.show()
```

```python
data_boxcox_diff = pd.Series(data_boxcox - data_boxcox.shift(), data_2.index)
plt.figure(figsize=(15,6))
plt.plot(data_boxcox_diff, label = 'difference after boxcox transformation')
plt.legend(loc='best')
plt.title('Difference after boxcox transformation\n', fontdict = {'fontsize':16, 'fontweight':5, 'color' : 'Orange'})
plt.show()
```

```python
data_boxcox_diff.dropna(inplace=True)
```

```python
adf_test = adfuller(data_boxcox_diff)
print('ADF_Statistics: %f' % adf_test[1])
print('Critical Values @ 0.05: %.2f' % adf_test[4]['5%'])
```

```python
train_data_boxcox = data_boxcox[:train_len]
test_data_boxcox =  data_boxcox[train_len:]
train_data_boxcox_diff = data_boxcox_diff[:train_len-1]
test_data_boxcox_diff =  data_boxcox_diff[train_len:]
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(data_boxcox_diff)
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(data_boxcox_diff, order=(2, 1, 9))
model_fit = model.fit()

import warnings
warnings.filterwarnings('ignore')
```

```python
y_hat_ar = data_boxcox_diff.copy()
y_hat_ar['ar_forecast_boxcox_diff']=model_fit.predict(data_boxcox_diff.index.min(), data_boxcox_diff.index.max())
y_hat_ar['ar_forecast_boxcox'] = y_hat_ar['ar_forecast_boxcox_diff'].cumsum()
y_hat_ar['ar_forecast_boxcox'] = y_hat_ar['ar_forecast_boxcox'].add(data_boxcox[0])
y_hat_ar['ar_forecast'] = np.exp(y_hat_ar['ar_forecast_boxcox'])
```

```python
plt.figure(figsize=(12,4))
plt.plot(train['General index'], label='train')
plt.plot(train['General index'], label='test')
plt.plot(y_hat_ar['ar_forecast'][test.index.min():], label='Auto regression forecast')
plt.legend(loc='best')
plt.title('Auto Regression Method')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(test['General index'], y_hat_ar['ar_forecast'][test.index.min():])).round(2)
mape = np.round(np.mean(np.abs(test['General index']-y_hat_ar['ar_forecast'][test.index.min():])/test['General index'])*100,2)
print(mape)
```
