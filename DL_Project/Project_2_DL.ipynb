```python
datagen = ImageDataGenerator()
```

```python
train_it = datagen.flow_from_directory('/content/dataRice/train/',
target_size=(224, 224),class_mode='categorical'
, batch_size=1)
```

```python
val_it = datagen.flow_from_directory('/content/dataRice/val/',
target_size=(224, 224),class_mode='categorical'
, batch_size=1)
```

```python
test_it = datagen.flow_from_directory('/content/dataRice/test/',target_size=(224,
224),class_mode='categorical', batch_size=1)
```

```python
base_model = VGG16(weights="imagenet", include_top=False, input_shape=(224,224,3),classes=5)
base_model.trainable = False
opt=Adam(learning_rate=0.0001)
model=Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(512, activation = 'relu'))
model.add(Dropout(0.25))
model.add(Dense(5,activation='softmax'))
model.compile(loss = keras.losses.categorical_crossentropy, optimizer=opt, metrics= ['accuracy'])
```

```python
h=model.fit_generator(train_it,epochs=5, validation_data=val_it)
```
